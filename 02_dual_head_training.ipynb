```python
import torch
from torch.utils.data import DataLoader
from datasets.joint_dataset import JointLungDataset
from models.dual_head_model import DualHeadModel
from evaluation.metrics import evaluate_model

DEVICE = "cuda"

# Load best model
model = DualHeadModel(feature_dim=256).to(DEVICE)
model.load_state_dict(torch.load("weights/best_dual_head.pth"))

val_dataset = JointLungDataset("processed_data/val_joint.pt", augment=False)
val_loader = DataLoader(val_dataset, batch_size=12, shuffle=False)

results = evaluate_model(model, val_loader, device=DEVICE)

for k, v in results.items():
    print(k, ":", v)
```
